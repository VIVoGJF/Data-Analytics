## How are in-demand skills trending for Data Analysts?

### Impoting and Cleaning Data

```python
#importing libraries
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
datasets = load_dataset('lukebarousse/data_jobs')
df = datasets['train'].to_pandas()

# data cleaning
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda skl_lst: ast.literal_eval(skl_lst) if pd.notna(skl_lst) else skl_lst )
```

### Group the Skills by Month

```python
df_india = df[(df.job_country == 'India') & (df.job_title_short == 'Data Analyst')].copy()

df_india['month_no'] = df_india['job_posted_date'].dt.month
df_india_explode = df_india.explode('job_skills')

df_india_pivot = df_india_explode.pivot_table(index='month_no', columns='job_skills',  aggfunc='size', fill_value=0)

df_india_pivot
```

#### Sorting by Skill Count

```python
df_india_pivot.loc['Total'] = df_india_pivot.sum()
df_india_pivot = df_india_pivot[df_india_pivot.loc['Total'].sort_values(ascending=False).index]
df_india_pivot = df_india_pivot.drop('Total')

df_india_pivot
```

### Plot the Monthly Skill Counts for Top 5 Skills

```python
df_india_pivot.iloc[:, :5].plot(kind='line')

plt.title('Trending Top Skills for Data Analysts in the India')
plt.ylabel('Count')
plt.ylim(0,500)
plt.xlabel('')
plt.legend()
plt.show()
```

### Calculate Percentage of Total Jobs

#### Getting monthly totals

```python
m_total = df_india.groupby('month_no').size()

m_total
```

#### Converting Skill Count to Percentage and Month Number to Month Name

```python
df_india_percent = df_india_pivot.iloc[:12].div(m_total/100, axis=0)


df_india_percent = df_india_percent.reset_index()
df_india_percent['month'] = df_india_percent['month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_india_percent = df_india_percent.set_index('month')
df_india_percent = df_india_percent.drop(columns='month_no')

df_india_percent
```

### Plot Monthly Skill Demand for Top 5 Skills

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_india_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in India')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```
